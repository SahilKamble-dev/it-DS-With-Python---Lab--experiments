1 :- What is need for Hypothesis Testing ?

Ans :- The need for Hypothesis Testing arises when we want to make decisions or draw conclusions about a population based on sample data. In statistics, it provides a structured way to test assumptions using evidence rather than guesswork.

Here are the main reasons why hypothesis testing is important:
1. Decision Making

Hypothesis testing helps in making logical decisions based on data.
Example: A company can decide whether a new product is better than the old one.
2. Testing Assumptions

It allows us to check whether an assumption (hypothesis) about a population is true or not.

Null hypothesis (H₀): No effect or no difference
Alternative hypothesis (H₁): There is an effect or difference

3. Scientific Research

In fields like Statistics and Data Science, hypothesis testing is essential for validating theories and experiments.
4. Reducing Uncertainty

Instead of relying on intuition, hypothesis testing uses probability to reduce uncertainty and provide more reliable conclusions.
5. Quality Control

Industries use it to check product quality and maintain standards.
Example: Testing whether a batch of products meets required specifications.
6. Comparing Groups

It helps compare two or more groups to see if differences are significant.
Example: Comparing student performance under two teaching methods.

7. Supports Data-Driven Decisions

It forms the basis of modern analytics and business intelligence, ensuring decisions are backed by data.

3 :- What is a student t distribution?

Ans :- Student’s t-distribution is a continuous probability distribution used in statistical analysis when the sample size is small and the population standard deviation is unknown.

It was developed by William Sealy Gosset.

Formula

t=s/n

Characteristics:

It is bell-shaped and symmetrical.

It has heavier tails than the normal distribution.

It depends on degrees of freedom (df = n − 1).

As sample size increases, it approaches the normal distribution.

Uses:

Used for small sample sizes (n < 30)

Applied when population standard deviation is unknown

Used in hypothesis testing (t-test)

Helps in confidence interval estimation

Conclusion:

Student’s t-distribution is an important statistical tool used to make reliable inferences about a population when only a small sample is available.

4 :- What is level of significance ? What role t plays in hypothesis Testing


Ans :- The level of significance (α) is the probability of rejecting the null hypothesis (H₀) when it is actually true.

In simple terms, it is the maximum acceptable error in hypothesis testing.
Common Values:

a = 0.05(5%)

a = 0.01(1%)

Importance:

It helps in decision-making during hypothesis testing.

It defines the critical region (rejection region).

It controls the Type I error (false rejection of H₀).

Role of t in Hypothesis Testing

The t-value comes from the Student’s t-distribution, developed by William Sealy Gosset.

The t-statistic is used to test hypotheses about the population mean when:

Sample size is small

```python
import numpy as np
from scipy import stats

# Sample data: ages of TE IT students
ages = [18, 19, 20, 19, 18, 21, 19, 20, 18, 19]

# Given population mean
mu = 19

# Perform one-sample t-test
t_statistic, p_value = stats.ttest_1samp(ages, mu)

print("t-statistic:", t_statistic)
print("p-value:", p_value)

# Level of significance
alpha = 0.05

# Decision
if p_value < alpha:
    print("Reject H0: Average age is not 19")
else:
    print("Fail to Reject H0: Average age is 19")
```
