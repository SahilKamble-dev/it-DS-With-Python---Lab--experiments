2 :- Explain what is Type I and Type II errors with an example.


Ans :- Type I and Type II Errors

In hypothesis testing, errors may occur while making decisions about the null hypothesis (H₀). These errors are classified as Type I error and Type II error.
1) Type I Error (α Error)

Definition:
Type I error occurs when the null hypothesis (H₀) is rejected even though it is true.

Symbol: α (level of significance)

Explanation:

It means we conclude that there is a significant effect when actually there is none.

Example:

Let,
H₀: Average age of students = 19

If the true average age is 19, but we reject H₀, then it is a Type I error.

2) Type II Error (β Error)

Definition:
Type II error occurs when the null hypothesis (H₀) is not rejected even though it is false.

Symbol: β


Explanation:
It means we fail to detect a real effect.


Example:
Let,
H₀: Average age of students = 19

If the actual average age is not 19, but we accept H₀, then it is a Type II error.


Real-life example:
A medical test shows a person is healthy, but actually, the person has a disease.

3 :- What is power of statistical test?


Ans :- Power of Statistical Test

Definition:
The power of a statistical test is the probability of rejecting the null hypothesis (H₀) when it is actually false.

Formula:
Power = 1 − β
(where β = probability of Type II error)

Explanation:

Power measures the ability of a test to correctly detect a real effect.
A higher power means the test is more effective in identifying false null hypotheses.


Example:

Let,
H₀: Average marks of students = 60

If the actual average marks are not 60, and the test correctly rejects H₀, then the test has high power.

Conclusion:

A good statistical test should have high power, meaning it should minimize Type II error (β).

4 :-  Explain input and out put characteristics for  stats.ttest_rel function in your own words


Ans :- Input and Output Characteristics of stats.ttest_rel()

The stats.ttest_rel() function (from the SciPy library) is used to perform a paired t-test. It compares two related samples to check whether their means are significantly different.

Input Characteristics

1 :- Two Related Samples (a, b):

The function takes two arrays/lists of data.

Both samples must be dependent (paired).

Example: marks of the same students before and after an exam.

2:- Same Size Requirement:

Both input samples must have equal number of observations.

3 :- Numeric Data:

Inputs should contain numerical values (integers or floats).

4 :- Optional Parameters:

axis: defines how the test is applied across data.

nan_policy: handles missing values (ignore, raise error, etc.).

alternative: specifies hypothesis type (two-sided, greater, less).

Output Characteristics

The function returns two main values:

1 :- t-statistic:

Measures the difference between paired samples.

Larger value ⇒ greater difference.

2 :- p-value:

Helps in decision making.

If p-value < significance level (α) → Reject H₀

If p-value ≥ α → Accept H₀

Conclusion

stats.ttest_rel() takes paired numerical data as input and gives t-statistic and p-value as output to determine whether there is a significant difference between the two related samples.


5 :- What is function in stats module for identifying power of a statistical test. Explain out put range of it and its significance.

Ans :- Function Description

ttest_power() is used to compute the power of a t-test, i.e., the probability of correctly rejecting the null hypothesis (H₀) when it is false.

Output Range
The output of ttest_power() lies between 0 and 1.

Interpretation:

Power = 0 → Test never detects a true effect (very poor test)
Power = 1 → Test always detects a true effect (ideal test)
Power ≥ 0.8 → Considered a good and reliable test

Significance of Power
Measures Test Effectiveness:
Higher power means the test is better at detecting real differences.

Reduces Type II Error (β):
Power = 1 − β, so higher power means lower chance of missing a true effect.

Helps in Sample Size Planning:
Researchers use power to decide the required sample size for accurate results.

Improves Decision Making:
Ensures that important effects are not ignored.

Conclusion

The ttest_power() function helps evaluate how strong a statistical test is. A higher power indicates a more reliable test with a lower probability of Type II error.
