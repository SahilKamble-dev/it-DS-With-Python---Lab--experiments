1 :- Explain input and output characteristics for f_oneway function from stats library

Ans :- Input and Output Characteristics of stats.f_oneway()

The f_oneway() function is used to perform a one-way ANOVA test, which compares the means of two or more independent groups to determine if at least one group mean is significantly different.

Input Characteristics

1 :- Two or More Independent Samples:
Each input is a 1-D array or list of numerical data.
Example: Scores of students in three different teachin methods.

2:- Sample Size:
Each sample can have a different number of observations.

3:- Numeric Data Only:
Data should be numbers (integers or floats).

4 :- Independent Groups:
Groups should be independent, meaning no repeated measurements between groups.

Syntax Example:

stats.f_oneway(sample1, sample2, sample3)

Output Characteristics

The function returns an object containing two values:

1:- F-statistic:
Measures the ratio of variance between groups to variance within groups.
Higher F → greater difference between group means.
 2 :- p-value:
 Probability of observing the data if H₀ is true.
Decision rule:
If p < α → Reject H₀ (at least one group mean is different)
If p ≥ α → Accept H₀ (no significant difference)
Example Output:
F = 5.32, p = 0.012

3 :- What role ANOVA plays in data analytics?


Ans :- Role of ANOVA in Data Analytics

ANOVA (Analysis of Variance) is a statistical method used to compare the means of two or more groups and determine whether the differences are statistically significant.

Roles in Data Analytics

1:- Compare Multiple Groups at Once:

ANOVA allows comparison of more than two groups simultaneously without increasing Type I error.
Example: Comparing sales performance across three regions.

2:- Identify Significant Factors:

Helps to find which categorical variables significantly influence a numerical outcome.
Example: Effect of teaching method on student scores.

3:- Supports Decision Making:

Provides F-statistic and p-value to guide whether observed differences are meaningful.
Helps analysts make data-driven decisions.

4:- Reduces Need for Multiple t-Tests:

Instead of performing multiple pairwise t-tests, ANOVA tests all groups in a single test, saving time and avoiding error inflation.

5 :- Explains Variability in Data:

ANOVA breaks down total variability into:
Between-group variability → due to factor differences
Within-group variability → random error

5:- Used in Predictive Modeling:

ANOVA results can inform feature selection in regression or classification models by identifying significant categorical features.

Example

In marketing analytics, ANOVA can be used to check if average purchase amounts differ across customer age groups.

H₀: No difference between age groups
H₁: At least one group differs
Result: F-statistic and p-value indicate whether age significantly affects purchase behavior.

Conclusion

ANOVA is a powerful tool in data analytics for:

Comparing multiple groups
Identifying significant factors
Guiding business or research decisions


4 :- When to use one way ANOVA? Explain with real world case.


Ans :- One-Way ANOVA
When to Use One-Way ANOVA

One-way ANOVA is used when you want to compare the means of three or more independent groups to check if there is a statistically significant difference among them based on one categorical factor.

Requirements:

One categorical independent variable (factor) with ≥ 3 groups
One numerical dependent variable
Observations are independent
Variances of groups are approximately equal

Real-World Example

Scenario:
A company wants to check if average productivity differs based on work shift (Morning, Afternoon, Night).

Independent variable (factor): Work Shift (3 categories)
Dependent variable: Productivity (number of units produced per worker)

Hypotheses:

H₀: Mean productivity is the same across all shifts
H₁: At least one shift has different mean productivity

Steps:

Collect productivity data for workers in each shift
Apply one-way ANOVA
Check F-statistic and p-val

Interpretation:

If p < 0.05, reject H₀ → At least one shift differs in productivity
If p ≥ 0.05, accept H₀ → No significant difference among shifts


Conclusion

One-way ANOVA is ideal for comparing means of three or more independent groups on a single factor and is widely used in business, education, healthcare, and research for decision-making based on group comparisons.


5 :-  When to use 2 way ANOVA ? Explain with real world case.


Ans :- Two-Way ANOVA
When to Use Two-Way ANOVA

Two-way ANOVA is used when you want to analyze the effect of two independent categorical factors on a numerical dependent variable and also see if there is an interaction effect between the two factors.

Requirements:

Two categorical independent variables (factors)
One numerical dependent variable
Observations are independent
Variances across groups are approximately equal
Real-World Example

Scenario:

A school wants to study how teaching method and class type affect student exam scores.

Factor 1: Teaching Method (Traditional, Online)
Factor 2: Class Type (Morning, Afternoon)
Dependent Variable: Student Scores

Hypotheses:

H₀₁: No difference in scores based on teaching method
H₀₂: No difference in scores based on class type
H₀₃: No interaction between teaching method and class type

Steps:

Collect student scores for each combination of teaching method and class type
Apply two-way ANOVA
Check F-statistics and p-values for each factor and interaction

Interpretation:

If p < 0.05 for a factor → That factor significantly affects scores
If p < 0.05 for interaction → Effect of one factor depends on the level of the other factor

Conclusion

Two-way ANOVA is ideal for examining two factors simultaneously, identifying main effects and interaction effects, and is widely used in education, business, manufacturing, and healthcare for multi-factor analysis.
