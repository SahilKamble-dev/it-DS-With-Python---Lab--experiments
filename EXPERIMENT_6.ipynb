Answer following question

1 :- Describe all three data set in terms of following : how many features, their categories (num,text,categorical, nominal,ordinal etc), number of tuples, missing values.


Ans :- To describe three datasets, we analyze:

Number of features
Feature categories (Numerical, Categorical, Nominal, Ordinal, Text)
Number of tuples
Missing values

```python
import pandas as pd

# ---------------- Dataset 1 ----------------
data1 = {
    "Age": [20, 21, 22, None],
    "Gender": ["Male", "Female", "Male", "Female"],
    "Salary": [20000, 25000, None, 30000],
    "City": ["Mumbai", "Pune", "Delhi", "Mumbai"],
    "Feedback": ["Good", "Excellent", "Average", "Good"]
}
df1 = pd.DataFrame(data1)

# ---------------- Dataset 2 ----------------
data2 = {
    "Marks": [80, 90, 85, None],
    "Grade": ["A", "A+", "B", "A"],
    "Subject": ["Math", "Science", "English", "Math"],
    "Remarks": ["Good", "Very Good", "Average", "Excellent"]
}
df2 = pd.DataFrame(data2)

# ---------------- Dataset 3 ----------------
data3 = {
    "Product_ID": ["P1", "P2", "P3", "P4"],
    "Price": [100, 200, 150, None],
    "Quantity": [2, 3, 1, 5],
    "Category": ["Electronics", "Clothing", "Electronics", "Clothing"],
    "Rating": [5, 4, None, 3],
    "Review": ["Good", "Nice", "Average", "Excellent"]
}
df3 = pd.DataFrame(data3)


# Function to analyze dataset
def analyze(df, name):
    print("\nDataset:", name)
    print("Features:", df.shape[1])
    print("Tuples:", df.shape[0])

    print("\nFeature Types:")
    for col in df.columns:
        if df[col].dtype in ['int64', 'float64']:
            print(col, "-> Numerical")
        else:
            print(col, "-> Categorical/Text")

    print("\nMissing Values:")
    print(df.isnull().sum())


# Analyze all datasets
analyze(df1, "Dataset 1")
analyze(df2, "Dataset 2")
analyze(df3, "Dataset 3")
```

3 :- In  Air Quality data set which graph is used to perform multivariate analysis?


Ans :- The Air Quality dataset contains multiple variables such as pollutants (CO, NO₂, SO₂), temperature, humidity, and wind speed.
To analyze the relationship between more than two variables simultaneously, multivariate analysis is performed.
1. Pair Plot (Scatter Plot Matrix)

The most commonly used graph for multivariate analysis is the Pair Plot.

 Explanation:
It displays scatter plots between every pair of variables
Diagonal shows distribution (histogram/KDE)
Helps to:
Identify correlation between variables
Detect patterns and trends
Find outliers

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create sample Air Quality dataset
data = {
    "CO": [2.3, 1.8, 2.5, 3.0, 2.2],
    "NO2": [30, 25, 40, 50, 35],
    "Temperature": [25, 27, 30, 35, 28],
    "Humidity": [60, 55, 50, 45, 65]
}

df = pd.DataFrame(data)

# Pair Plot
sns.pairplot(df)
plt.show()
```

Heatmap

```python
sns.heatmap(df.corr(), annot=True)
plt.show()
```

3D Scatter Plot

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df['CO'], df['Temperature'], df['Humidity'])
plt.show()
```

4 :- In Wine quality data set can you do non graphical EDA ? Explain your answer.


Ans :- Non-Graphical EDA on Wine Quality Dataset

The Wine Quality dataset contains chemical properties of wine such as:

Fixed acidity
Volatile acidity
Citric acid
pH
Alcohol
Quality (output variable)

Techniques Used in Non-Graphical EDA
1. Summary Statistics
Mean
Median
Standard deviation
Minimum and Maximum

```python
import pandas as pd

# Create sample Wine dataset
data = {
    "fixed_acidity": [7.4, 7.8, 7.8, 11.2, 7.4],
    "volatile_acidity": [0.70, 0.88, 0.76, 0.28, 0.70],
    "citric_acid": [0.00, 0.00, 0.04, 0.56, 0.00],
    "pH": [3.51, 3.20, 3.26, 3.16, 3.51],
    "alcohol": [9.4, 9.8, 9.8, 9.8, 9.4],
    "quality": [5, 5, 5, 6, 5]
}

df = pd.DataFrame(data)

# 1. Summary statistics
print(df.describe())

# 2. Missing values
print(df.isnull().sum())

# 3. Data types
print(df.dtypes)

# 4. Quality distribution
print(df['quality'].value_counts())

# 5. Correlation
print(df.corr())

# 6. Skewness
print(df.skew())
```
