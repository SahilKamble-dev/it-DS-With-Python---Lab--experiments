1 :- When do we go for Chi square test ?  

Ans :- Chi-Square Test
When do we use Chi-Square Test?

The Chi-square test is used when we want to analyze categorical (non-numerical) data and check whether there is a significant relationship or difference between variables.

Situations to Use Chi-Square Test
Testing Independence of Variables:
To check whether two categorical variables are related or independent.
Example: Relationship between gender and product preference.
Goodness of Fit Test:
To check whether observed data fits an expected distribution.
Example: Whether dice outcomes follow equal probability.
Comparing Proportions:
To compare frequencies or counts in different categories.
Example: Number of students choosing different streams.

Conditions for Applying Chi-Square Test
Data should be in frequency (count) form
Categories should be mutually exclusive
Sample size should be large enough (expected frequency ≥ 5 preferred)
Observations should be independent

Example

Suppose we want to check whether smoking habit and health condition are related:

H₀: No relationship between smoking and health
H₁: There is a relationship

Conclusion

We use the Chi-square test when dealing with categorical data to test independence, association, or goodness of fit between variables.

2 :- Do we use Chi square test for independence ? Explain How ?


Ans :- Chi-Square Test for Independence
Do we use Chi-square test for independence?

Yes, the Chi-square test is used to check whether two categorical variables are independent or associated.

Explanation

The test compares the observed frequencies (O) with the expected frequencies (E) under the assumption that the variables are independent.

Steps Involved
Form Hypotheses:
H₀: The two variables are independent
H₁: The two variables are dependent (associated)
Create Contingency Table:
Arrange data in rows and columns (e.g., Gender vs Preference)
Calculate Expected Frequencies:

E=(Row Total×Column Total)/Grand Total

Compute Chi-Square Statistic:
χ2=∑E(O−E)^2/E
Decision Rule:
Compare calculated χ² value with table value OR
Use p-value approach
If p-value < α → Reject H₀

Example

To check whether gender and product choice are independent:

Collect data → form contingency table
Apply Chi-square formula
If H₀ is rejected → variables are dependent

Conclusion

The Chi-square test for independence helps determine whether two categorical variables are related or independent by comparing observed and expected frequencies.

3 :- Explain role of Critical value and p-value in Chi square test.


Ans :- Role of Critical Value and p-value in Chi-Square Test

In a Chi-square test, both critical value and p-value are used to decide whether to accept or reject the null hypothesis (H₀).
1) Role of Critical Value

Definition:
The critical value is the threshold value obtained from the Chi-square table based on:

Level of significance (α)
Degrees of freedom (df)

Role in Decision Making:

If calculated χ² > critical value → Reject H₀
If calculated χ² ≤ critical value → Accept H₀

2) Role of p-value

Definition:
The p-value is the probability of obtaining the observed result assuming H₀ is true.

Role in Decision Making:

If p-value < α → Reject H₀
If p-value ≥ α → Accept H₀

Conclusion

Both critical value and p-value help in hypothesis testing:

Critical value gives a cut-off point
p-value gives the exact significance level

4 :- Can we use chi square test to show existence of correlation between 2 features in a dataset ?  How will you interpret test result to indicate correlation presence.


Ans :- Using Chi-Square Test to Show Correlation Between Two Features
1 :- ) Can we use it?

Yes, the Chi-square test of independence can be used to check whether there is a relationship (association/correlation) between two categorical features in a dataset.
2) How to Use It

Step 1: Form Hypotheses

H₀: The two features are independent (no correlation)
H₁: The two features are dependent (correlated/associated)

Step 2: Create Contingency Table

Arrange the frequency of observations for each combination of feature categories.

Step 3: Compute Chi-Square Statistic

χ2=∑E(O−E)^2​/E
O = observed frequency
E = expected frequency (assuming independence)
Step 4: Compare with Critical Value or p-value

If χ² > critical value OR p-value < α → Reject H₀ → Features are associated
If χ² ≤ critical value OR p-value ≥ α → Accept H₀ → Features are independent
3) Interpretation
Reject H₀ → There is evidence of correlation (association) between the two features.
Accept H₀ → No evidence of correlation; the features are independent.

Example:
Check if Gender and Product Choice are related:

H₀: Gender and Product Choice are independent
H₁: Gender and Product Choice are associated

If test result:

p-value = 0.02 and α = 0.05 → Reject H₀ → Gender influences product choice

Conclusion

Chi-square test can indicate the presence of correlation between categorical features by testing whether the variables are independent or associated.

5 :- What is degrees of freedom and how to calculate it while performing chi square test.


Ans :- Degrees of Freedom in Chi-Square Test

1) Definition

Degrees of freedom (df) in a Chi-square test refers to the number of independent values or categories that can vary while performing the test.

It is used to determine the critical value from the Chi-square table.
Higher df → more categories → wider distribution.

2) How to Calculate

a) For Chi-square Test of Independence (Contingency Table):
df=(Number of rows−1)×(Number of columns−1)
Rows: Number of categories of first variable
Columns: Number of categories of second variable
Example:
A table of Gender (2 categories) vs Product Choice (3 categories)
df=(2−1)×(3−1)=1×2=2

b) For Chi-square Goodness of Fit Test:
df=(Number of categories−1−number of parameters estimated)
Usually, no parameters estimated → df = categories − 1

3) Role in Chi-Square Test
Determines the shape of the Chi-square distribution.
Used to find critical value for decision-making.
Ensures the test accounts for independent information only.

Conclusion

Degrees of freedom (df) quantify the number of independent
comparisons in the data and are essential for interpreting the
Chi-square statistic correctly.
